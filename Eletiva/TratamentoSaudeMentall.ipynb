```python
import pandas as pd
import numpy as np
```

```python
# Carrega o arquivo CSV com os dados de saúde mental
saude_mental_base = pd.read_csv(r'C:\Users\kauea\OneDrive\Área de Trabalho\ERROR 404\Faculdade\Eletiva\Student_Mental_Health_Clean.csv')
saude_mental = saude_mental_base
saude_mental
```

```python
# Renomeia as colunas para nomes em português
saude_mental.rename(columns={
    "Timestamp": "data_hora",
    "Choose your gender": "genero",
    "Age": "idade",
    "What is your course?": "curso",
    "Your current year of Study": "ano_estudo",
    "What is your CGPA?": "rendimento_academico",
    "Marital status": "estado_civil",
    "Do you have Depression?": "tem_depressao",
    "Do you have Anxiety?": "tem_ansiedade",
    "Do you have Panic attack?": "tem_ataque_panico",
    "Did you seek any specialist for a treatment?": "procurou_especialista",
    "Sleep Hours": "horas_sono",
    "Exercise Frequency": "frequencia_exercicio",
    "Social Support": "suporte_social",
}, inplace=True)
```

```python
# Traduz valores binários (Yes/No) para Sim/Não
binarias = ["tem_depressao", "tem_ansiedade", "tem_ataque_panico", "procurou_especialista","suporte_social"]
for col in binarias:
    saude_mental[col] = saude_mental[col].map({"Yes": "Sim", "No": "Não"})

# Traduz gênero para português
saude_mental["genero"] = saude_mental["genero"].map({
    "Male": "Masculino",
    "Female": "Feminino"
})

# Padroniza o texto do ano de estudo
saude_mental["ano_estudo"] = saude_mental["ano_estudo"].str.lower().str.replace("year", "ano").str.strip()

# Traduz os nomes dos cursos para português
saude_mental["curso"] = saude_mental["curso"].replace({
    "Engineering": "Engenharia",
    "Engine": "Engenharia",
    "engin": "Engenharia",
    "Educação Islâmica": "Educação Islâmica",
    "Islamic Education": "Educação Islâmica",
    "Pendidikan Islam": "Educação Islâmica",
    "Pendidikan Islam ": "Educação Islâmica",
    "BIT": "Tecnologia da Informação",
    "IT": "Tecnologia da Informação",
    "CTS": "Tecnologia da Informação",
    "Direito": "Direito",
    "Law": "Direito",
    "Mathemathics": "Matemática",
    "Matemática": "Matemática",
    "Ciência da Computação": "Ciência da Computação",
    "Biomedical science": "Ciência Biomédica",
    "Ciência Biomédica": "Ciência Biomédica",
    "Human Resources": "Recursos Humanos",
    "Psychology": "Psicologia",
    "psychology": "Psicologia",
    "Human Sciences ": "Ciências Humanas",
    "Communication ": "Comunicação",
    "Radiography": "Radiologia",
    "Biotechnology": "Biotecnologia",
    "Marine science": "Ciências Marinhas",
    "Accounting ": "Contabilidade",
    "Banking Studies": "Estudos Bancários",
    "Business Administration": "Administração de Empresas",
    "Usuluddin ": "Teologia Islâmica (Usuluddin)",
    "Fiqh": "Jurisprudência Islâmica (Fiqh)",
    "Fiqh fatwa ": "Jurisprudência Islâmica (Fiqh)",
    "ALA": "Língua Árabe Avançada",
    "TAASL": "Ensino de Árabe como Segunda Língua",
    "BENL": "Língua Inglesa para Negócios",
    "Benl": "Língua Inglesa para Negócios",
    "ENM": "Economia e Negócios",
    "KOE": "Engenharia (Faculdade de Engenharia - KOE)",
    "Koe": "Engenharia (Faculdade de Engenharia - KOE)",
    "koe": "Engenharia (Faculdade de Engenharia - KOE)",
    "KIRKHS": "Ciências Humanas Islâmicas (KIRKHS)",
    "Kirkhs": "Ciências Humanas Islâmicas (KIRKHS)",
    "Irkhs": "Conhecimento e Civilização Islâmica (IRKHS)",
    "KENMS": "Economia e Ciências de Gestão (KENMS)",
    "MHSC": "Ciências da Saúde",
    "Malcom": "Comunicação Multimídia",
    "Kop": "Farmácia",
    "DIPLOMA TESL": "Diploma em Ensino de Inglês como Segunda Língua",
    "Diploma Nursing": "Técnico em Enfermagem"
})

# Traduz faixas de rendimento acadêmico
saude_mental["rendimento_academico"] = saude_mental["rendimento_academico"].replace({
    "3.00 - 3.49": "3,00 - 3,49",
    "3.50 - 4.00": "3,50 - 4,00"
})

# Traduz frequência de exercício para português
saude_mental["frequencia_exercicio"] = saude_mental["frequencia_exercicio"].replace({
    "Never": "Nunca",
    "1-2 times": "1 a 2 vezes",
    "3-4 times": "3 a 4 vezes",
    "5+ times": "5 ou mais vezes",
    "Everyday": "Diariamente"
})
```

```python
# Função para classificar a área do conhecimento com base no nome do curso
def classificar_area(curso):
    curso = str(curso).lower()

    if any(palavra in curso for palavra in ["engenharia", "matemática", "ciência da computação", "tecnologia", "informação", "estatística"]):
        return "Exatas"
    elif any(palavra in curso for palavra in ["psicologia", "direito", "ciências humanas", "educação", "administração", "contabilidade", "economia", "comunicação", "recursos humanos", "islâmica", "jurisprudência", "língua", "teologia", "negócios"]):
        return "Humanas"
    elif any(palavra in curso for palavra in ["biomédica", "biologia", "marinhas", "farmácia", "saúde", "nursing", "radiologia", "biotecnologia"]):
        return "Biológicas"
    else:
        return "Outros"

# Aplica a classificação ao DataFrame
saude_mental["area_conhecimento"] = saude_mental["curso"].apply(classificar_area)

saude_mental
```
